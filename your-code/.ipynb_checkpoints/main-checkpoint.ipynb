#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data 


```python
import pymysql
import sqlalchemy 
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats)

```python
engine = sqlalchemy.create_engine('mysql+pymysql://guest:relational@relational.fit.cvut.cz:3306/stats')
```

#### 4. Import the users table 

```python
df_users = pd.read_sql_query('SELECT * FROM stats.users',engine)
```

```python
df_users.head()
```

#### 5. Rename Id column to userId

```python
df_users = df_users.rename(columns = {'Id':'userId'})
```

```python
df_users
```

#### 6. Import the posts table. 

```python
df_posts = pd.read_sql_query('SELECT * FROM posts',engine)
```

```python
df_posts.head().T
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
df_posts = df_posts.rename(columns = {'Id':'postId', 'OwnerUserId':'userId'})
df_posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
df_users.head()
```

```python
df_posts.head()
```

```python
df_users_n = df_users[['userId','Reputation','Views','UpVotes','DownVotes']]
df_users_n.head()
```

```python
df_posts_n = df_posts[['postId', 'Score','userId','ViewCount','CommentCount']]
df_posts_n.head()
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python

df_merged = pd.merge(df_users_n, df_posts_n, on = 'userId')
```

```python
df_merged.head(100)
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
df_outliers = df_merged.copy()
```

```python
df_outliers.info()
```

```python
df_outliers.describe(percentiles=[.25,.5,.75,.99]).T
```

```python
#there are not outliers at the lower limit

lower_limit = df_outliers.mean()-3*df_outliers.std()
lower_limit
```

```python
upper_limit = df_outliers.mean()+3*df_outliers.std()
upper_limit
```

```python
df_outliers[['Reputation', 'Views', 'UpVotes',	'DownVotes', 'Score', 'ViewCount', 'CommentCount']] = df_outliers[df_outliers > df_outliers.mean() + 3*df_outliers.std()][['Reputation', 'Views', 'UpVotes',	'DownVotes', 'Score', 'ViewCount', 'CommentCount']]
```
